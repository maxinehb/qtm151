### (a) Transform a list of values and plot it

- Create an array with the following values

- `vec_x = [-5,-3, 0, 10, 20, 40]`

- Create a new vector `vec_y` with the following formula

- $\qquad y = [cos(x)]^2$

- Produce a scatter plot of `vec_y` against `vec_x`
- Label the axes and the title
- Change the color to `red`, `green`, or `purple` (you choose)

<br>

- **Hint**: You can use the `numpy` functions

```python
#create arrays vec_x and vec_y
vec_x = np.array([-5,-3,0,10,20,40])
vec_y = np.cos(vec_x)**2

#produce scatter plot
plt.plot(vec_x, vec_y, color = "purple")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plot of X values against Y values")
plt.grid(True)
plt.show()
```

### (b) IF/ELSE statements

- Create two lists: 

- ` list_fruits = ["apple","orange","pear","banana"]`
- ` list_colors = ["blue","red","green","orange"]`

- Create a new variable ` message = "orange"`
- Create an if/else that tests whether ` message ` is a fruit and a color.  Print a message if the test is true.

<br>

- **Hint**: Use the `in` command to test whether an element is part of a list (see Lecture 4) 

```python
# create lists
list_fruits = ["apple","orange","pear","banana"]
list_colors = ["blue","red","green","orange"]

#new variable
message = "orange"

if (message in list_fruits) & (message in list_colors):
    print("The message is a fruit and a color!")
else:
    print("The message is not a fruit and a color.")
```

### (d) Create the following list of months:

` list_months = ["August","February"] ` 

` list_spring_semester = ["January", "February", "March", "April", "May"] `

Create a for-loop that checks whether each element of `list_months` belongs to the spring semester. Create a message with this result. 

The output generated by your loop should look like this:

`"August is NOT part of the Spring semester"`

`"February is part of the Spring semester"`

- **Hint**: Use if/else statements inside the loop. You can use the command "in" (see Lecture 4) to test whether an element belongs to a list


```python
# Write your answer here:
list_months = ["August", "February"]
list_spring_semester= ["January", "February", "March", "April", "May"]

for months in list_months:
    if months in list_spring_semester:
        print(months + " is part of the Spring semester")
    else:
        print(months + " is NOT part of the Spring semester")
```

### (e) Loop through dfferent variables and plot

- Read the dataset `data_quiz/wdi_ageprop_2020.csv`

` list_varnames = ["percent_ages0to14","percent_ages15to64"]`

- Run a for loop that creates a histogram plot for each of the variables on the list
- Label the axes (the y-axis can be called "Frequency")
- The loop should automatically create titles that look like this:
  - "Distribution for percent_ages0to14"
  - "Distribution for percent_ages15to64"

Note: This exercise tells you how much the proportion of the populaton in different ages varies across countries in 2020.

```python
# Write your answer here:
dataset = pd.read_csv('data_quiz/wdi_ageprop_2020.csv')

list_varnames = ["percent_ages0to14","percent_ages15to64"]


for names in list_varnames:
    plt.hist(dataset[names])
    plt.xlabel(names)
    plt.ylabel("Frequency")
    plt.title("Distribution for " + names)
    plt.show()


```
